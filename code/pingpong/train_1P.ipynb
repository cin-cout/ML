```python
# 

scene_info=[]
command=[]
for i in range(0,len(data)):
    scene_info=np.concatenate([scene_info,data[i]['ml_1P']['scene_info'][1:-20]],axis=0)

i=0
k = range(1, len(scene_info)-1)
ball_x = np.array([scene_info[i]['ball'][0] for i in k])
ball_y = np.array([scene_info[i]['ball'][1] for i in k])
ball_speed_x = np.array([scene_info[i+1]['ball'][0] - scene_info[i]['ball'][0] for i in k])
ball_speed_y = np.array([scene_info[i+1]['ball'][1] - scene_info[i]['ball'][1] for i in k])
direction = np.where(np.vstack((ball_speed_x, ball_speed_y)) > 0, [[1],[0]], [[2],[3]]).sum(axis=0)  # x y: ++1, +-4, -+2, --3
platform = np.array([scene_info[i]['platform_1P'][0] for i in k])
blocker_x=np.array([scene_info[i]['blocker'][0] for i in k])
blocker_speed_x = np.array([scene_info[i+1]['blocker'][0] - scene_info[i]['blocker'][0] for i in k])

```

```python

X = np.hstack((ball_x.reshape(-1, 1),
               ball_y.reshape(-1, 1),
               ball_speed_x.reshape(-1, 1),
               ball_speed_y.reshape(-1, 1),
               direction.reshape(-1, 1),
             blocker_x.reshape(-1, 1),
             blocker_speed_x.reshape(-1, 1)))

y=np.hstack(platform.reshape(-1,1))
```

```python
X, y
```

```python
# train data
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=1,weights='distance',p=1)
print(model.fit(X, y))
print(model.score(X, y))
```

```python
with open('../ml/model_1P.pickle', 'wb') as f:
    pickle.dump(model, f)
```
