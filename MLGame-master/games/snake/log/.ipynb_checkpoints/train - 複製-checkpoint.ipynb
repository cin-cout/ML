```python
# beta 8.0.1
import pickle
import os.path
import numpy as np
dirs=os.listdir(os.getcwd())
data=[]
for i in dirs:
    
    if 'pickle' and 'ml' in i:
        with open(i, 'rb') as f:
            data.append(pickle.load(f))
        

print(data[0].keys())
```

```python
for i in range(0,len(data)):
    data[i]['ml'].keys()
```

```python
# 提取特徵

scene_info=[]
command=[]
for i in range(0,len(data)):
    scene_info=np.concatenate([scene_info,data[i]['ml']['scene_info']],axis=0)
    command=np.concatenate([command,data[i]['ml']['command']],axis=0)
    
i=1
k = range(1, len(scene_info))

snake_head_x = np.array([scene_info[i]['snake_head'][0] for i in k])
snake_head_y = np.array([scene_info[i]['snake_head'][1] for i in k])
snake_speed_x = np.array([scene_info[i]['snake_head'][0] - scene_info[i-1]['snake_head'][0] for i in k])
snake_speed_y = np.array([scene_info[i]['snake_head'][1] - scene_info[i-1]['snake_head'][1] for i in k])

target = np.where(np.array(command) == 'RIGHT', 0,
                  np.where(np.array(command) == 'LEFT', 1, 
                  np.where(np.array(command) == 'UP', 2,
                  np.where(np.array(command) == 'DOWN', 3, 4))))[1:]
```

```python
X = np.hstack((snake_head_x.reshape(-1, 1),
               snake_head_y.reshape(-1, 1))

y = target
```

```python
print(X)
```

```python
print(y)
```

```python
# train data
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=1)
print(model.fit(X, y))
print(model.score(X, y))
```

```python
with open('../ml/modelfordraw.pickle', 'wb') as f:
    pickle.dump(model, f)
```
